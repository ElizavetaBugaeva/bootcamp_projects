```python
#sample link to song: href="/lyric/32105073/David+Bowie/Boys+Keep+Swinging
song_pattern = '/lyric/+\d+\S*"'
```

Find all link extentions for each song


```python
gorilazz_songs_ext= re.findall(song_pattern,gorillaz_html)
```

```python
len(gorilazz_songs_ext)
```

Define function which will exlude all dupliates from the songs list


```python
def clean_list(list):
    avoid = ['Version', 'Remastered', 'Live', 'DVD', 'RMX', 'Instrumental', 'Mix', 'Studio', '5D', 'Rmx','Soundtrack']
    return [elem for elem in list if not any(word in elem for word in avoid)]
```

```python
g_songs_cleanedup= clean_list(gorilazz_songs_ext)
```

```python
len(g_songs_cleanedup)
```

```python
g_songs_cleanedup
```

Save txt files for all gorillaz songs

```python
#remove all songs with duplicate titles using set
seen_titles = set()

for counter, link in enumerate(g_songs_cleanedup):
    url = 'https://www.lyrics.com' + link
    song_title = link.split('/')[4].rstrip("\"")
    if song_title not in seen_titles:
        response = requests.get(url, headers=headers)
        with open(f'gorillaz/{song_title}.txt', 'w', encoding='utf-8') as f:
            f.write(response.text)
        print(counter, song_title, url)
        seen_titles.add(song_title)
```
