## Extract the URLs

```python
song_pattern = '/lyric/+\d+\S*"'
```

### David Bowie

```python
david_songs = re.findall(song_pattern,david_b_html)
david_songs
```

```python
for counter, link in enumerate(david_songs):
    url = 'https://www.lyrics.com' + link
    response = requests.get(url)
    song_title = link.split('/')[4]
    with open(f'david_bowie/{song_title}.txt', 'w', encoding='utf-8') as f:
        f.write(response.text)
    print(counter, song_title, response.status_code)
```
