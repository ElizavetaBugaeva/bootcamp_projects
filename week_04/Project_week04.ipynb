Define function which will exlude all dupliates from the songs list


```python
def clean_list(list):
    avoid = ['Version', 'Remastered', 'Live', 'DVD', 'RMX', 'Instrumental', 'Mix', 'Studio', '5D', 'Rmx','Soundtrack']
    return [elem for elem in list if not any(word in elem for word in avoid)]
```

### Save txt files

```python
def file_creator(songs_cleanedup,dir):
    seen_titles = set()
    for counter, link in enumerate(songs_cleanedup):
        url = 'https://www.lyrics.com' + link
        song_title = link.split('/')[4].rstrip("\"")
        if song_title not in seen_titles:
            response = requests.get(url, headers=headers)
            with open(f'{dir}/{song_title}.txt', 'w', encoding='utf-8') as f:
                f.write(response.text)
            print(counter, song_title, url)
            seen_titles.add(song_title)
```

### Create lists of lines out of the created txt files

```python
def get_all_lines(dir_path, all_lines):
    all_lines = []
    for filename in os.listdir(dir_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(dir_path, filename)
            with open(file_path, 'r', encoding='utf-8') as f: 
                file_soup = BeautifulSoup(f, 'html.parser')
                lyrics_elem = file_soup.find('pre', class_="lyric-body")
                if lyrics_elem is not None:
                    for line in lyrics_elem.get_text().split('\n'):
                        line = line.strip("\r")  # remove carriage return
                        if len(line) > 0:
                            all_lines.append(line)
    return all_lines
```
