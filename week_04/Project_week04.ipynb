Define function which will exlude all dupliates from the songs list


```python
def clean_list(list):
    avoid = ['Version', 'Remastered', 'Live', 'DVD', 'RMX', 'Instrumental', 'Mix', 'Studio', '5D', 'Rmx','Soundtrack']
    return [elem for elem in list if not any(word in elem for word in avoid)]
```

```python
all_gorillaz_lines = []

for filename in os.listdir(gorillaz_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(gorillaz_path, filename)
        with open(file_path, 'r', encoding='utf-8') as f: 
            gorillaz_soup = BeautifulSoup(f, 'html.parser')
            lyrics_elem = gorillaz_soup.find('pre', class_="lyric-body")
            if lyrics_elem is not None:
                for line in lyrics_elem.get_text().split('\n'):
                    line = line.strip("\r")  # remove carriage return
                    if len(line) > 0:
                        all_gorillaz_lines.append(line)
```

### Vectorise text with Bag of words

Use the CountVectorizer from Scikit to transform your corpus into a matrix.


```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(
    lowercase = True, 
    analyzer = 'word',
    stop_words = 'english',
    token_pattern = r"(?u)\b\w\w+\b",
    ngram_range = (1, 1)
)
```

```python
vec = count_vectorizer.fit_transform(CORPUS)
```

```python
count_matrix_skl = pd.DataFrame(
    vec.todense(), 
     columns=count_vectorizer.get_feature_names(),
    index=LABELS
)
```

```python
count_matrix_skl
```

Normalize the counts using the TfidfTransformer.


```python
tf = TfidfTransformer(use_idf=True)
```

```python
vec2 = tf.fit_transform(vec)
```
